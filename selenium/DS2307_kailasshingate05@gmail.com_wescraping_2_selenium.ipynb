# Q1: Write a python program to scrape data for ‚ÄúData Analyst‚Äù Job position in ‚ÄúBangalore‚Äù location. You
have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10
jobs data.
This task will be done in following steps:
1. First get the webpage https://www.shine.com/
2. Enter ‚ÄúData Analyst‚Äù in ‚ÄúJob title, Skills‚Äù field and enter ‚ÄúBangalore‚Äù in ‚Äúenter the location‚Äù field.
3. Then click the searchbutton.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.


```python
#loading page
driver=webdriver.Edge()
driver.get('https://www.shine.com/')
#maximizing window
driver.maximize_window()
time.sleep(10)

 #clicking cancle button on popup
x=driver.find_element(By.CLASS_NAME,"btn-close")
x.click()
#loading search tag
button = driver.find_element(By.XPATH,'//input[@class="input"]')
time.sleep(5)
button.click()
time.sleep(10)
#sending keys to search tag
search_button=driver.find_element(By.XPATH,"//input[@id='id_q']")
search_button.send_keys("Data Analyst")

#sending loc to loc search tag
search_loc=driver.find_element(By.XPATH,"//input[@id='id_loc']")
search_loc.send_keys("Bangalore")
time.sleep(5)
#hitting the search button
search_button=driver.find_element(By.XPATH,"//button[text()='Search jobs']")
search_button.click()
#loading job titles
job_title=[i.text for i in(driver.find_elements(By.XPATH,'//h2[@itemprop="name"]'))[0:10]]
#loading job loc
job_loc=[re.findall(r'[A-za-z]+',i.text.strip()) for i in(driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]'))[0:10]]
#loading company name
company_name=[i.text for i in(driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]'))[0:10]]
#loading experiance
exp=[i.text for i in(driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]'))[0:10]]
#creating dataframe
df=pd.DataFrame({'Job Title':job_title,'Location':job_loc,'Company':company_name,'Experiance':exp})
df

```

# Q4: Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:
6. Brand
7. ProductDescription
8. Price
The attributes which you have to scrape is ticked marked in the below image.
To scrape the data you have to go through following steps:
1. Go to Flipkart webpage by url :https://www.flipkart.com/
2. Enter ‚Äúsunglasses‚Äù in the search fieldwhere ‚Äúsearch for products, brands and more‚Äù is written and
click the search icon
3. After that you will reach to the page having a lot of sunglasses. From this page you can scrap the
required data as usual.
4. After scraping data from the first page, go to the ‚ÄúNext‚Äù Button at the bottom other page , then
click on it.
5. Now scrape data from this page as usual
6. Repeat this until you get data for 100sunglasses

```python
#loading webpage
driver=webdriver.Edge()
driver.get("https://www.flipkart.com/")
driver.maximize_window()
#canceling login popup
try:
    driver.find_element(By.XPATH,'//button[@class="_2KpZ6l _2doB4z"]').click()
    #sending item name to search tag
    driver.find_element(By.XPATH,'//input[@class="_3704LK"]').send_keys('Sunglasses')
    driver.find_element(By.XPATH,'//button[@class="L0Z3Pu"]').click()
    time.sleep(2)
except:
     pass
#iterating pages
brand=[]
desc=[]
price=[]
discount=[]
for i in range(3):
        #loading brand names
        b=driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
        for i in b:
            brand.append(i.text)
        #loading description
        d=driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
        for i in d:
              desc.append(i.text)
        #loading price of sunglasses
        p=driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
        for i in p:
            price.append(i.text)
        #loading discount on sunglasses
        dc=driver.find_elements(By.XPATH,'//div[@class="_3Ay6Sb"]')
        for i in dc:
            discount.append(i.text)
        driver.find_element(By.XPATH,'//span[text()="Next"]').click()
        time.sleep(5)
brand=[brand[i]for i in range(0,100)]
price=[price[i]for i in range(0,100)]
desc=[desc[i]for i in range(0,100)]
discount=[discount[i]for i in range(0,100)]

#creating DataFrame
df=pd.DataFrame({'Brand':brand,'Description':desc,'Price':price,'Discount%':discount})
df

```

# Q5: Scrape 100 reviews data from flipkart.com for iphone11 phone. You have to go the link:
https://www.flipkart.com/apple-iphone-11-black-64-gb/productreviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART
As shown in the above page you have to scrape the tick marked attributes. These are:
1. Rating
2. Review summary
3. Full review
4. You have to scrape this data for first 100reviews.

```python
#loading webpage
driver=webdriver.Edge()
driver.get('https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&market')
#extracting rating 
rating=[]
Review_summary=[]
Full_review=[]
#iterating pages
for i in range(10):
    #extracting rating 
    rat_tag=driver.find_elements(By.XPATH,'//div[@class="_3LWZlK _1BLPMq"]')
    for tag in rat_tag:
        rating.append(tag.text)
    #loading summary
    rev_tag=driver.find_elements(By.XPATH,'//p[@class="_2-N8zT"]')
    for tag in rev_tag:
        Review_summary.append(tag.text)
    #loading full summary
    sum_tag=driver.find_elements(By.XPATH,'//div[@class="t-ZTKy"]')
    for tag in sum_tag:
        Full_review.append(tag.text)
    #loading next page    
    driver.find_element(By.XPATH,'//span[text()="Next"]').click()
    time.sleep(2)
    
rating=[rating[i] for i in range(100)]
Review_summary=[Review_summary[i] for i in range(100)]
Full_review=[Full_review[i] for i in range(100)] 
#creating DataFrame
df=pd.DataFrame({'Rating':rating,'Review':Review_summary,'Full Review':Full_review})
df
```

# Q8: Write a python program to scrape data for Top 1000 Quotes of All Time.
The above task will be done in following steps:
1. First get the webpagehttps://www.azquotes.com/
2. Click on TopQuotes
3. Than scrap a) Quote b) Author c) Type Of Quotes

```python
#load webpage
driver=webdriver.Edge()
driver.get('http://www.azquotes.com/')
#full screen mode
driver.maximize_window()
#cliclking on top quotes
driver.find_element(By.XPATH,'/html[1]/body[1]/div[1]/div[1]/div[1]/div[1]/div[3]/ul[1]/li[5]/a[1]').click()

for i in range(10):
#getting quote
    quoteTage=driver.find_elements(By.XPATH,'//a[@class="title"]')
    quote=[]
    for i in quoteTage:
        quote.append(i.text)
    #gettign authore details
    authoreTag=driver.find_elements(By.XPATH,'//div[@class="author"]')
    authore=[]
    for i in authoreTag:
        authore.append(i.text)
    #extracting type of quotes details    
    quoteTage=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    quoteType=[]
    for i in quoteTage:
        quoteType.append(i.text)
    #navigate to next page
    driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[3]/li[12]/a').click()
#creating dataframe
df=pd.DataFrame({"Quote":quote,'Authore':authore,"Type of Quote":quoteType})
df
```

# Q9: Write a python program to display list of respected former Prime Ministers of India(i.e. Name, Born-Dead,Term of office, Remarks) from https://www.jagranjosh.com/.
This task will be done in following steps:
1. First get the webpagehttps://www.jagranjosh.com/
2. Then You have to click on the GK option
3. Then click on the List of all Prime Ministers of India
4. Then scrap the mentioned data and make theDataFrame

```python
#loading webpage
driver=webdriver.Edge()
driver.get('https://www.jagranjosh.com/')
driver.maximize_window()
#navigating on gk page
driver.find_element(By.XPATH,"//a[text()='GK']").click()
#navigating to list of prime minister 
time.sleep(5)
driver.find_element(By.XPATH,"//a[text()='List of Prime Ministers of India ']").click()

# driver.find_element(By.XPATH,'/html/body/div/div/div[1]/div[1]/div/svg/path[1]').click()
time.sleep(5)
name=[]
BornDead=[]
Termofoffice=[]
Remark=[]
for i in range(2,21):
    n_tag=driver.find_element(By.XPATH,'//*[@id="itemdiv"]/div[4]/span/div[2]/table/tbody/tr['+str(i)+']/td[2]')
    name.append(re.sub(r'\(Acting\)','',str(n_tag.text)))
    #loading born_dead data    
    b_tag=driver.find_element(By.XPATH,'//*[@id="itemdiv"]/div[4]/span/div[2]/table/tbody/tr['+str(i)+']/td[3]')
    BornDead.append(b_tag.text)
    #loadinf Termofoffice data 
    t_tag=driver.find_element(By.XPATH,'//*[@id="itemdiv"]/div[4]/span/div[2]/table/tbody/tr['+str(i)+']/td[4]')
    Termofoffice.append(t_tag.text)
    #loading Remarks
    r_tag=driver.find_element(By.XPATH,'//*[@id="itemdiv"]/div[4]/span/div[2]/table/tbody/tr['+str(i)+']/td[5]')
    Remark.append(r_tag.text)
#creating dataframe
df=pd.DataFrame({"PM Name":name,"Born-Dead":BornDead,"Term of office":Termofoffice,"Remarks":Remark})
df
```

# Q10: Write a python program to display list of 50 Most expensive cars in the world (i.e.Car name and Price)
from https://www.motor1.com/
This task will be done in following steps:
1. First get the webpage https://www.motor1.com/
2. Then You have to type in the search bar ‚Äô50 most expensive cars‚Äô
3. Then click on 50 most expensive carsin the world..
4. Then scrap the mentioned data and make the dataframe

```python
#loading webpage
driver=webdriver.Chrome()
driver.get('https://uk.motor1.com/')
driver.maximize_window()
#switching region
driver.find_element(By.XPATH,'//div[@class="m1-editions-switcher"]').click()
time.sleep(2)
driver.find_element(By.XPATH,'/html/body/div[10]/div[1]/div[4]/ul/li[1]/a').click()
#finding search option and sending keys
search_bar=driver.find_element(By.XPATH,'//input[@id="search_input"]')
search_bar.send_keys("50 most expensive cars")
time.sleep(2)
driver.find_element(By.XPATH,"//button[@class='m1-search-panel-button m1-search-form-button-animate icon-search-svg']//*[name()='svg']").click()
#clicking on link
button = driver.find_element(By.XPATH,"//a[normalize-space()='50 Most Expensive Cars In The World']").click()
#loading car name
car_name=[]
car_name_tag=driver.find_elements(By.XPATH,'//h3[@class="subheader"]')
for i in car_name_tag[0:50]:
    car_name.append(i.text)

#loading price details
car_price=[]
for i in range(4,103,2):
    car_price_tag=driver.find_element(By.XPATH,'//*[@id="article_box"]/div[1]/div[2]/div[1]/p['+str(i)+']/strong')
    car_price.append(re.findall(r'\$\d+\.\d\s+\w+',str(car_price_tag.text)))

#creating DataFrame
df=pd.DataFrame({"Car Name":car_name,"Car Price":car_price})
df
```
