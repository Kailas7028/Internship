# 1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url = 
https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: 
A) Rank 
B) Name 
C) Artist 
D) Upload date 
E) Views 

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
from selenium.common.exceptions import NoSuchElementException,ElementNotInteractableException,TimeoutException,StaleElementReferenceException
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
```

```python
#loading driver
driver=webdriver.Chrome()
#loading page
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
wait=WebDriverWait(driver,10)
#declaring lists
rank=[]
name=[]
artist=[]
updateDate=[]
views=[]
#closing popup
try:
    wait.until(EC.element_to_be_clickable((By.XPATH,'//div[@class="frb-step frb-step-1"]/button[1]'))).click()
except:
    pass
#getting rank 
ranktag=wait.until(EC.presence_of_all_elements_located((By.XPATH,'//div[@id="mw-content-text"]/div/table[2]/tbody/tr/td[1]')))
for i in ranktag:
    rank.append(i.text)
#getting video names
nametag=wait.until(EC.presence_of_all_elements_located((By.XPATH,'//div[@id="mw-content-text"]/div/table[2]/tbody/tr/td[2]')))
for i in nametag:
    name.append(i.text)
#getting artist names
artisttag=wait.until(EC.visibility_of_all_elements_located((By.XPATH,'//div[@id="mw-content-text"]/div/table[2]/tbody/tr/td[3]')))
for i in artisttag:
    artist.append(i.text)
#getting update date 
datetag=wait.until(EC.visibility_of_all_elements_located((By.XPATH,'//div[@id="mw-content-text"]/div/table[2]/tbody/tr/td[4]')))
for i in datetag:
    updateDate.append(i.text)
viewstag=wait.until(EC.presence_of_all_elements_located((By.XPATH,'//div[@id="mw-content-text"]/div/table[2]/tbody/tr/td[5]')))
for i in ranktag:
    views.append(i.text)
driver.quit()
print(len(rank),len(name),len(artist),len(updateDate),len(views))
```

```python
#creating dataframe
df=pd.DataFrame({"Rank":rank,"Video Name":name,"Artist":artist,"Update Date":updateDate,"Views":views})
df
```
